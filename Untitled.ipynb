```python
"""
Task 1
"""

# generate a vector of random numbers which obeys the given distribution.
#
# n: length of the vector
# mu: mean value
# sigma: standard deviation.
# dist: choices for the distribution, you need to implement at least normal 
#       distribution and uniform distribution.
#
# For normal distribution, you can use ``numpy.random.normal`` to generate.
# For uniform distribution, the interval to sample will be [mu - sigma/sqrt(3), mu + sigma/sqrt(3)].

import numpy as np
import math
from math import sqrt
import random
import matplotlib.pyplot as plt
from matplotlib import pyplot as plt
from matplotlib import animation

def generate_random_numbers(n, mu, sigma, dist="normal"):
    # write your code here.
    
    if dist == "normal":
        return np.random.normal(mu, sigma, n)
    elif dist == "uniform":
        return np.random.normal(mu-sigma/math.sqrt(3),mu+sigma/math.sqrt(3),n)
    else:
        raise Exception("The distribution {unknown_dist} is not implemented".format(unknown_dist=dist))
        
        
# test your code:
y_test = generate_random_numbers(5, 0, 0.1, "normal")
print(y_test)
```

```python
y1 = generate_random_numbers(105, 0.5, 1.0, "normal")
y2 = generate_random_numbers(105, 0.5, 1.0, "uniform")
```

```python
# IGD, the ordering is permitted to have replacement. 
#
#

def IGD_wr_task1(y):
    n = len(y)
    ordering = np.random.choice(n, n, replace=True)
    x0=0
    k=0
    xk=np.zeros(n)
    obj=np.zeros(n)
    xk[0]=x0
    while k+1<=n:
        gamma=1/(k+1)
        obj[k]=1/2*np.sum((xk[k-1]-y)**2)
        xk[k]=xk[k-1]-gamma*(xk[k-1]-y[ordering[k]])
        #obj[k]=1/2*np.sum((xk[k]-y)**2)
        k+=1
    return xk[-1], obj
    # implement the algorithm's iteration of IGD. Your result should return the the final xk
    # at the last iteration and also the history of objective function at each xk.


# IGD, the ordering is not permitted to have replacement.
#
#
def IGD_wo_task1(y):
    n = len(y)
    ordering = np.random.choice(n, n, replace=False)
    x0=0
    k=0
    xk=np.zeros(n)
    obj=np.zeros(n)
    xk[0]=x0
    while k+1<=n:
        gamma=1/(k+1)
        obj[k]=1/2*np.sum((xk[k-1]-y)**2)
        xk[k]=xk[k-1]-gamma*(xk[k-1]-y[ordering[k]])
        #obj[k]=1/2*np.sum((xk[k]-y)**2)
        k+=1
    return xk[-1], obj
    # implement the algorithm's iteration of IGD. Your result should return the the final xk
    # at the last iteration and also the history of objective function at each xk.
    
xk1, obj1 = IGD_wr_task1(y1)
xk2, obj2 = IGD_wo_task1(y1)

histories1  = obj1
histories2 = obj2

for i in range(20):
    histories1 += IGD_wr_task1(y1) [1]
    histories2 += IGD_wo_task1(y1)[1]
    
histories1 = histories1 / 21
histories2 = histories2 / 21


plt.figure()
plt.title('normal distrubution')
with_rep = plt.plot(np.log(histories1),c = 'red',label = 'with replacement')
without_rep = plt.plot(np.log(histories2),c = 'blue',label = 'without replacement')
plt.legend()

xk3, obj3 = IGD_wr_task1(y2)
xk4, obj4 = IGD_wo_task1(y2)

histories3 = obj3
histories4 = obj4

for i in range(20):
    histories3 += IGD_wr_task1(y2) [1]
    histories4 += IGD_wo_task1(y2)[1]
    
histories3 = histories3 / 21
histories4 = histories4 / 21

plt.figure()
plt.title('uniform distrubution')
with_rep = plt.plot(np.log(histories3),c = 'red',label = 'with replacement')
without_rep = plt.plot(np.log(histories4),c = 'blue',label = 'without replacement')
plt.legend() 
```

```python
'''
Task 3
'''

# generation of exact solution and data y and matrix A.

def generate_problem_task3(m, n, rho):
    A = np.random.normal(0., 1.0, (m, n))
    x = np.random.random(n) # uniform in (0,1)
    w = np.random.normal(0., rho, m)
    y = A@x + w
    return A, x, y

# We generate the problem with 200x100 matrix. rho as 0.01.
#
A, xstar, y = generate_problem_task3(200, 100, 0.01)

# In these two functions, we could only focus on the first n steps and try to make comparisons on these data only.
# In practice, it requires more iterations to converge, due to the matrix might not be easy to deal with.
# You can put the ordering loop into a naive loop: namely, we simply perform the IGD code several rounds.
#
#
#
# IGD, the ordering is permitted to have replacement. 
#
#
def IGD_wr_task3(y, A):
    # implement the algorithm's iteration of IGD. Your result should return the the final xk
    # at the last iteration and also the history of objective function at each xk.
    gamma = 1e-3
    n = A.shape[0]
    ordering = np.random.choice(n, n, replace=True)
    xk = np.zeros((n,A.shape[1]))
    obj = np.zeros(n)
    x0=np.zeros(A.shape[1])
    xk[0]=x0
    k=0
    while k+1 <= n :
        obj[k] = np.sum((np.matmul(A,xk[k-1]) - y)**2)
        xk[k,:] = xk[k-1] - gamma * A[ordering[k],:]* (np.matmul(A[ordering[k],:],xk[k-1]) - y[ordering[k]])
        k += 1
    return xk, obj


# IGD, the ordering is not permitted to have replacement.
#
#
def IGD_wo_task3(y, A):
    # implement the algorithm's iteration of IGD. Your result should return the the final xk
    # at the last iteration and also the history of objective function at each xk.
    gamma = 1e-3
    n = A.shape[0]
    ordering = np.random.choice(n, n, replace=False)
    xk = np.zeros((n,A.shape[1]))
    obj = np.zeros(n)
    x0=np.zeros(A.shape[1])
    xk[0]=x0
    k=0
    while k+1 <= n :
        obj[k] = np.sum((np.matmul(A,xk[k-1]) - y)**2)
        xk[k,:] = xk[k-1] - gamma * A[ordering[k],:] * (np.matmul(A[ordering[k],:],xk[k-1]) - y[ordering[k]])
        k += 1
    return xk, obj
```

```python
from numpy import linalg as LA
xk1, obj1 = IGD_wr_task3(y,A)
xk2, obj2 = IGD_wo_task3(y,A)

histories1  = obj1
histories2 = obj2

for i in range(20):
    histories1 += IGD_wr_task3(y,A)[1]
    histories2 += IGD_wo_task3(y,A)[1]
    
histories1 = histories1 / 21
histories2 = histories2 / 21

plt.figure()
plt.title('value of objective function')
with_rep = plt.plot(histories1,c = 'red',label = 'with replacement')
without_rep = plt.plot(histories2,c = 'blue',label = 'without replacement')
plt.legend()


xk1_conv_rate = LA.norm((xk1-xstar),axis = 1)
xk2_conv_rate = LA.norm((xk2-xstar),axis = 1)


histories3  = xk1_conv_rate
histories4 = xk2_conv_rate

for i in range(80):
    histories3 += LA.norm((xk2-xstar),axis = 1)[1]
    histories4 += LA.norm((xk2-xstar),axis = 1)[1]
    
histories3 = histories3 / 81
histories4 = histories4 / 81



plt.figure()
plt.title('convergence rate of x')
x_with_rep = plt.plot(histories3,c = 'red',label = 'with replacement')
x_without_rep = plt.plot(histories4,c = 'blue',label = 'without replacement')
plt.legend()
```
